```python
def read_file(filename):
    df = pd.read_csv(filename, parse_dates=[3], skipfooter=1)
    df.index = df.datetime
    return df
```

```python
df2 = read_file(filename = 'El Paso Chamizal/AQDM_2142091599.txt')
df2.describe()
```

```python
df2.value.plot()
```

```python
df1.value.corr(df2.value)
```
