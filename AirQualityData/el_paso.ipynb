```python
def read_file(filename):
    df = pd.read_csv(filename, parse_dates=[3])
    df.dropna(subset=['datetime'], inplace=True)
    df.index = df.datetime
    df.sort_index(inplace=True)
    return df
```

```python
from glob import glob

def get_filenames(dirname, pattern='*.txt'):
    return glob(dirname + '/' + pattern)
```

```python
dirname = 'El Paso Chamizal'

res = []
for filename in get_filenames(dirname):
    df = read_file(filename)
    if len(df) == 0:
        continue
    date = df.index[0]
    row = df.loc[date]
    if not np.isnan(row.parameter):
        res.append((row.parameter, date, filename))
    
res.sort()
res = pd.DataFrame(res)
res
```
